# Secure Coding Review

This project demonstrates how to identify and fix common security vulnerabilities in a Python login system.

```python
print("===== Welcome to ABC Login System =====")

username = input("Enter Username: ")
password = input("Enter Password: ")

stored_username = "admin"
stored_password = "admin123"

name = "Administrator"
email = "admin@gmail.com"
phone = "9876543210"

if username == stored_username:
    if password == stored_password:
        print("\nLogin Successful")
        print("Welcome", name)
        print("Email:", email)
        print("Phone:", phone)

        print("\nMenu")
        print("1. View Profile")
        print("2. Change Password")
        print("3. Logout")

        choice = input("Enter your choice: ")

        if choice == "1":
            print("\nProfile")
            print(name)
            print(email)
            print(phone)

        elif choice == "2":
            new_password = input("Enter New Password: ")
            stored_password = new_password
            print("Password Changed Successfully")

        elif choice == "3":
            print("Logged Out Successfully")

        else:
            print("Invalid Choice")

    else:
        print("Incorrect Password")

else:
    print("Username Not Found")
```

## Vulnerabilities Found

During the security review of the login system, the following vulnerabilities were identified:

1. Hardcoded Username and Password
2. Password Stored in Plain Text
3. Weak Password Policy
4. No Password Hashing
5. Unlimited Login Attempts
6. No Input Validation
7. Sensitive Information Disclosure
8. Password Change Without Old Password Verification

## Secure Login System

The following code is the secure version of the login system with improved security features.

import hashlib

print("===== Secure Login System =====")

# Stored user information
stored_username = "admin"

# Password Hash
stored_password_hash = hashlib.sha256("admin123".encode()).hexdigest()

name = "Administrator"
email = "admin@gmail.com"
phone = "9876543210"

# Maximum login attempts
max_attempts = 3
attempt = 0

while attempt < max_attempts:

    username = input("Enter Username: ")
    password = input("Enter Password: ")

    # Hash entered password
    entered_password_hash = hashlib.sha256(password.encode()).hexdigest()

    if username == stored_username and entered_password_hash == stored_password_hash:

        print("\nLogin Successful")
        print("-----------------------------")
        print("Welcome:", name)
        print("-----------------------------")

        while True:

            print("\nMenu")
            print("1. View Profile")
            print("2. Change Password")
            print("3. Logout")

            choice = input("Enter your choice: ")

            if choice == "1":

                print("\nProfile")
                print("Name :", name)
                print("Email:", email)
                print("Phone:", phone)

            elif choice == "2":

                old_password = input("Enter Old Password: ")

                old_hash = hashlib.sha256(old_password.encode()).hexdigest()

                if old_hash == stored_password_hash:

                    new_password = input("Enter New Password: ")

                    if len(new_password) >= 8:

                        stored_password_hash = hashlib.sha256(new_password.encode()).hexdigest()

                        print("Password Changed Successfully")

                    else:

                        print("Password should be at least 8 characters.")

                else:

                    print("Old Password Incorrect.")

            elif choice == "3":

                print("Logged Out Successfully")
                break

            else:

                print("Invalid Choice")

        break

    else:

        attempt += 1

        print(f"Invalid Credentials. Remaining Attempts: {max_attempts-attempt}")

if attempt == max_attempts:

    print("\nAccount Temporarily Locked.")
